```python

import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
#stemming

sno = nltk.stem.SnowballStemmer('english')
```

```python
print(sno.stem("operating"))
print(sno.stem("operated"))
print(sno.stem("operation"))
print(sno.stem("operator"))
print(sno.stem("casting"))
print(sno.stem("whatsup"))
print(sno.stem("advertising"))
```

```python
def data_cleaning(series):
    i=0
    string= ""
    final_string = []
    list_of_sent = []
    cleanr= re.compile('<.*?>')
    
    
    for sent in series.values:
        filtered_sent = []
        sent = re.sub(cleanr, ' ', sent) #searching and replacing all '<.*?>' with space
        sent = re.sub('[^a-zA-Z0-9\n]', ' ', sent)
        sent = re.sub('\s+', ' ', sent)
        sent = sent.lower()
        for word in sent.split():
            if word not in stop_words and len(word)>2:
                word = sno.stem(word)
                filtered_sent.append(word)
        list_of_sent.append(filtered_sent)
        string = " ".join(filtered_sent)
        final_string.append(string)
        i+=1
    return final_string, list_of_sent
                
        
```

```python

for x in final['Text'].iloc[:5].values:
    print(x,"\n\n")
```

```python
final_string, list_of_sent = data_cleaning(final['Text'].iloc[:5])
for x in final_string:
    print(x,"\n\n")
```

```python
start = time.time()
final_string, list_of_sent = data_cleaning(final['Text'])
end = time.time()
print("Time takes in seconds =", end - start)
```
